````python
# Rename columns to more user friendly names
# List format used for ease of reading
# inplace=True means the changes will be applied to the original dataframe
attendance.rename(columns={

    'Person Code' : 'Person Code',
    'Unit Instance Code' : 'Module Code',
    'Calocc Code' : 'Year',
    'Surname' : 'Surname',
    'Forename' : 'Forename',
    'Long Description' : 'Module Name',
    'Register Event ID' : 'Event ID',
    'Object ID' : 'Object ID',
    'Register Event Slot ID' : 'Event Slot ID',
    'Planned Start Date' : 'Date',
    'is Positive' : 'Has Attended',
    'Postive Marks' : 'Attended',
    'Negative Marks' : 'Not Attended',
    'Usage Code' : 'Attendance Code'

}, inplace=True)
````

````python
# Print updated columns to check changes have been made correctly (no typos)
print(attendance.columns)
````

5. Convert `Date` to a pandas timestamp. 
   > tip: use `pd.to_datetime()`

````python
# Check the current data type (dtype) of the 'Date' column
print(attendance['Date'].dtype)
````

````python
# Convert the 'Date' column from object dtype to datetime format
attendance['Date'] = pd.to_datetime(attendance['Date'])
# Datetime format allows for easier manipulation and analysis of data related to dates and times
````

````python
# Check the updated dtype of the 'Date' column
print(attendance['Date'].dtype)
````

6. Filter the DataFrame on **one specific module** (your choice) and plot its **attendance rate over time**
   (x = Date, y = the module's average attendance)

````python
# Print all unique values in the "Module Name" column, alongised their counts (from highest to lowest)
attendance["Module Name"].value_counts()
````

````python
# Create a new dataframe filtered to only include entries where the "Module Name" is "Algorithms"
# Algorithms selected as module since it has the highest number of entries
algorithms_attendance = attendance[attendance["Module Name"] == "Algorithms"] 
````

````python
# Using the column "Attended" columns ( 1/0 values) makes it easier to calculate and plot attendance rates
# Group by 'Date' and calculate the mean attendance for each date (proportion of students present each day)
algorithms_daily_attendance = algorithms_attendance.groupby('Date')['Attended'].mean().reset_index()
# reset_index() is used to convert this back into a normal DataFrame
````

````python
# Print the first few rows of the new dataframe to check
print(algorithms_daily_attendance.head())
````

````python
# Plot the average attendance over time for the Algorithms module
plt.figure(figsize=(10,5)) # set figure size to be wider for easier reading
plt.plot(algorithms_daily_attendance["Date"],
         algorithms_daily_attendance["Attended"],
         marker="o") # line plot with markers at each data point

plt.title("Average Attendance Over Time (Algorithms Module)")
plt.xlabel("Date")
plt.ylabel("Average Attendance Rate")
plt.xticks(rotation=45)  # rotate labels so theyâ€™re readable
plt.grid(True) # add grid for easier reading of values
plt.show()
````
